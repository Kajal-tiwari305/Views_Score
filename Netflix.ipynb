**MAKING PREDICTION USING LM10**

```python
X = MC[['weekend','Character_A','ad_impression_million']]
X = sm.add_constant(X)
Predicted_views = lm_10.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(MC.Views_show,Predicted_views)
r_squared = r2_score(MC.Views_show,Predicted_views)
```

```python
print('Mean_Squared_Error :', mse)
print('r_square_value:',r_squared)
```

```python
# actual VS Predicted 
c = [i for i in range(1,81,1)]
fig=plt.figure()
plt.plot(c,MC.Views_show, color='blue',linewidth=2.5 , linestyle='-' )
plt.plot(c,Predicted_views, color='red',linewidth=2.5 , linestyle='-' )
fig.suptitle('Actual and Prediction', fontsize=20)
plt.xlabel('Index', fontsize=18)
plt.xlabel('Index', fontsize=16)
```

```python
c=[i for i in range(1,81,1)]
fig = plt.figure()
plt.plot(c,MC.Views_show-Predicted_views, color='pink' ,linewidth=2.5,linestyle='-')
fig.suptitle('error terms',fontsize = 20)
plt.xlabel('Index', fontsize=18)                      # X-label
plt.ylabel('Views_show-Predicted_views', fontsize=16) 
```
